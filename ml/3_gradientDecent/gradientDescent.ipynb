```python
import matplotlib.pyplot as plt
def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.01
    plt.plot(x,y,color="red",marker="+")
    
    for i in range(iterations):
        y_predicted = [m_curr * x_val + b_curr for x_val in x]
        plt.plot(x,y_predicted,color="green")
        cost = (1/n) * sum((y[j] - y_predicted[j])**2 for j in range(n))
        md = -(2/n) * sum(x[j] * (y[j] - y_predicted[j]) for j in range(n))
        bd = -(2/n) * sum(y[j] - y_predicted[j] for j in range(n))
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        # print(f"m {m_curr}, b {b_curr}, cost {cost} iteration {i}")
# Sample data
x = [1, 2, 3, 4, 5]
y = [5, 7, 9, 11, 13]

gradient_descent(x, y)
```
