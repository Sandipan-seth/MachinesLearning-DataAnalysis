```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
import seaborn as sns
import math
df=pd.read_csv("canada_per_capita_income.csv")
df.tail(11)
```

```python
sns.lmplot(x='year', y='perCapitaIncome', data=df)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.plot(df.year,df.perCapitaIncome,color="red",marker="+")
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['year']],df['perCapitaIncome'])
value=reg.predict([[2020]])
print(math.floor(value))
```

```python
reg = linear_model.LinearRegression()

reg.fit(df[['year']], df['perCapitaIncome'])

year_2020 = pd.DataFrame({'year': [2020]})
print(year_2020)

value = reg.predict(year_2020)
print(value)

predicted_value = value[0] 
print(math.floor(predicted_value))
```
